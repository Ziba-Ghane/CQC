# creat_empty_bug_class_predictions

```python
def creat_empty_bug_class_predictions(df, l1_th):
    # df: DataFrame("bug_report_id", "label|prediction_th")
    
    gdf = df.groupby("bug_report_id")
    empty_df_bug_preds = None

    columns_name = ["bug_report_id"]
    columns_name.extend(my_bug_classes_config)
    columns_dtype = ["int64"]
    columns_dtype.extend(columns_dtype_config)
    print(columns_name)
    print(columns_dtype)

    # empty_df_bug_preds
    temp = {}
    for cn, cd in zip(columns_name, columns_dtype):
        temp[cn] = pd.Series(dtype=cd)
    empty_bug_class_predictions = pd.DataFrame(temp)

    if l1_th:
        empty_bug_class_predictions["bug_report_id"] = gdf.groups.keys()
        empty_bug_class_predictions["bug_class_2"] = empty_bug_class_predictions["bug_class_2"].astype("Int32")
    else:
        empty_bug_class_predictions["bug_class_4"] = pd.NA
        empty_bug_class_predictions["bug_report_id"] = gdf.groups.keys()
        empty_bug_class_predictions["bug_class_2"] = empty_bug_class_predictions["bug_class_2"].astype("Int32")
        empty_bug_class_predictions["bug_class_3"] = empty_bug_class_predictions["bug_class_3"].astype("Int32")
    
    return empty_bug_class_predictions
```

# plot_bug_class_distribution

```python
# --------------------------- bar plot

def plot_bug_class_distribution(my_bug_classes, num_bugs, each_bug_class_num, each_bug_class_per, bla, th):
    colours = ["red", "blue", "green", "cyan"]
    # figsize=(a, b) : a_width, b_height
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 10))
    width = 0.2
    my_x = [-0.3, -0.1, 0.1, 0.3]
    
    temp1 = str(num_bugs)
    temp2 = len(temp1) - 1
    temp3 = "1" + "0" * temp2
    temp4 = temp1[0] + "0" * temp2
    temp5 = int(temp3) + int(temp4)
    
    ml = list( range(0, temp5, int(temp3)) )
    ml.append(num_bugs)
    
    for i, key in enumerate(my_bug_classes):
        ax1 = axes[i][0]
        ax2 = axes[i][1]
        for j, cl in enumerate(each_bug_class_per[key]):
            ax1.bar(my_x[j], each_bug_class_num[key][cl], width=width, label=cl, color=colours[j])
            ax2.bar(my_x[j], each_bug_class_per[key][cl], width=width, label=cl, color=colours[j])
        
        tempXlabel = "{}".format(key)
        ax1.set_xlabel(tempXlabel)
        ax2.set_xlabel(tempXlabel)
        
        ax1.set_ylabel("Number of Bug Reports")
        ax2.set_ylabel("Percentages (%)")
        
        tempXlim = [-0.5, 0.5]
        ax1.set_xlim(tempXlim)
        ax2.set_xlim(tempXlim)
        
        
        ax1.set_ylim([0, temp5])
        ax2.set_ylim([0, 110])
        
        
        ax1.get_xaxis().set_ticks([])
        ax2.get_xaxis().set_ticks([])
        
        
        ax1.get_yaxis().set_ticks(ml)
        ax2.get_yaxis().set_ticks(list(range(0, 101, 10)))
        
        ax1.legend()
        ax2.legend()
        
        ax1.grid(axis="y")
        ax2.grid(axis="y")
    
    fig.suptitle("bla: {}, th: {}".format(bla, th), fontsize=16)
    fig.tight_layout(pad=3.0)
```
