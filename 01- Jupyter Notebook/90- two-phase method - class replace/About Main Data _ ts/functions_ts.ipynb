# prediction_column_name

```python
def prediction_column_name(th):
    return "predictions_" + str(th)
```

# create_empty_unique_class_predictions

```python
def create_eucp_columns_name_dtype():
    columns_name = preprocessing_params["output"]["unique_class_predictions"]["columns_name"]
    columns_dtype = preprocessing_params["output"]["unique_class_predictions"]["columns_dtype"]

    temp_column_number = len(columns_dtype)
    for th in knn_accuracy_threshold:
        temp_column_name = prediction_column_name(th)
        columns_name.append(temp_column_name)
        columns_dtype[temp_column_number] = pd.Int32Dtype()
        temp_column_number += 1
    return columns_name, columns_dtype
```

```python
def create_empty_unique_class_predictions():
    # columns name & columns dtype
    columns_name, columns_dtype = create_eucp_columns_name_dtype()

    # create df structure
    temp_df_structure = {}
    for cn, cd in zip(columns_name, columns_dtype.values()):
        temp_df_structure[cn] = pd.Series(dtype=cd)

    # create: empty_unique_class_predictions
    empty_unique_class_predictions = pd.DataFrame(temp_df_structure)

    return empty_unique_class_predictions
```

# percentage_predictable_classes

```python
def calc_percentage_predictable_classes(unique_class_predictions):
    num_total = len(unique_class_predictions)
    per_class_predictable = []
    for th in knn_accuracy_threshold:
        new_column_name = prediction_column_name(th)
        temp2 = num_total - unique_class_predictions[new_column_name].isna().sum()
        temp3 = round((temp2 / num_total) * 100)
        per_class_predictable.append(temp3)
    return per_class_predictable
```

# plot_percentage_predictable_classes

```python
def plot_percentage_predictable_classes(percentage_predictable_classes):
    # figsize=(a, b) : a_width, b_height
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))

    axes.set_title(software_name)
    axes.set_xlabel("KNN threshold accuracy")
    axes.set_ylabel("percentage of pridictable classes")

    tempp = list( map(str, knn_accuracy_threshold) )
    axes.set_xticks(range(len(knn_accuracy_threshold)))
    axes.set_xticklabels(tempp)

    axes.set_ylim([0, 110])
    axes.get_yaxis().set_ticks(list(range(0, 101, 10)))
    
    axes.grid(True)
    axes.plot(range(len(knn_accuracy_threshold)), percentage_predictable_classes, label="kl", color="blue", 
                linestyle="solid", lw=1, marker=".")
```
