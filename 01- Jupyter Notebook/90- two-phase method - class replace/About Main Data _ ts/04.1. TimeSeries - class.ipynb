# Config

```python
mypaths = {}
for software_name, v in dataset_file_name_ts.items():
    mypaths[software_name] = project_folder + data_folder + software_name + "/" + v + ".csv"
```

```python
preprocessing_params = {
    "data": {
        "dataset_timeseries": {
            "columns_name":   ["class_id", "time_stamp", "label"],
            "columns_dtype" : {0: "int64", 1: "int64", 2:"int32"}
        }
    }
}
```

# Read TimeSeries

```python
for k, v in mypaths.items():
    df_timeseries = pd.read_csv(
        v, 
        names=preprocessing_params["data"]["dataset_timeseries"]["columns_name"], 
        dtype=preprocessing_params["data"]["dataset_timeseries"]["columns_dtype"], 
        header=None
    )
    print("{}:".format(k), len(df_timeseries))
```
